```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
%matplotlib inline
```

```python
covid_cases = pd.read_csv("counties.csv")
```

```python
msno.matrix(covid_cases.sample(250))
```

```python
null_pattern = (np.random.random(1000).reshape((50, 20)) > 0.5).astype(bool)
null_pattern = pd.DataFrame(null_pattern).replace({False: None})
msno.matrix(null_pattern.set_index(pd.period_range('1/1/2011', '2/1/2015', freq='M')) , freq='BQ')
```

```python
msno.bar(covid_cases.sample(1000))
```

```python
msno.heatmap(covid_cases)
```

```python
msno.dendrogram(covid_cases)
```
