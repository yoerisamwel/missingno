```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
%matplotlib inline
```

```python
data_set = pd.read_csv("Seattle_Housing_Cost_Burden.csv")
```

```python
msno.matrix(data_set.sample(250))
```

```python
null_pattern = (np.random.random(650).reshape((13, 50)) > 0.5).astype(bool)
null_pattern = pd.DataFrame(null_pattern).replace({False: None})
msno.matrix(null_pattern.set_index(pd.period_range('1/1/2006', '2/1/2018', freq='y')) , freq='BQ')
```

```python
msno.bar(data_set.sample(1000))
```

```python
msno.heatmap(data_set)
```

```python
msno.dendrogram(data_set)
```
